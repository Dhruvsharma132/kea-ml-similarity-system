```python
# Step 1: Imports
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Step 2: Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Step 3: Sample inputs
inputs = [
    "I need help generating leads",
    "Looking for an AI chatbot for my website",
    "Want to improve conversion rates",
    "Need help building a sales funnel",
    "Interested in marketing automation tools"
]

# Step 4: Query
query = "I want to get more leads for my business"

# Step 5: Encode text → vectors
input_embeddings = model.encode(inputs)
query_embedding = model.encode([query])

# Step 6: Compute similarity
similarity_scores = cosine_similarity(query_embedding, input_embeddings)[0]

# Step 7: Find best match
best_index = np.argmax(similarity_scores)
best_match = inputs[best_index]
best_score = similarity_scores[best_index]

# Step 8: Output
print("🔍 Query:", query)
print("✅ Best Match:", best_match)
print("📊 Similarity Score:", round(float(best_score), 3))
```

```python
def get_top_k_matches(query, inputs, k=3):
    input_embeddings = model.encode(inputs)
    query_embedding = model.encode([query])

    scores = cosine_similarity(query_embedding, input_embeddings)[0]

    ranked = sorted(
        zip(inputs, scores),
        key=lambda x: x[1],
        reverse=True
    )

    return ranked[:k]


top_matches = get_top_k_matches(query, inputs)

print("\nTop Matches:")
for i, (text, score) in enumerate(top_matches, 1):
    print(f"{i}. {text} (Score: {round(float(score), 3)})")
```
