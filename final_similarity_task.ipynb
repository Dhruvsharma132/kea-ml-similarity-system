# 🔍 Semantic Similarity System (KeaBuilder ML Task)

## Objective
Match user inputs (leads/prompts) with the most relevant existing entries using semantic similarity.

## Approach
- Convert text to embeddings using Sentence Transformers
- Compute cosine similarity
- Return best and top-k matches

## Use Case
- Lead matching
- Template recommendations
- Query clustering

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
def find_best_match(query, inputs):
    input_embeddings = model.encode(inputs)
    query_embedding = model.encode([query])

    scores = cosine_similarity(query_embedding, input_embeddings)[0]
    best_index = np.argmax(scores)

    return {
        "query": query,
        "best_match": inputs[best_index],
        "score": float(scores[best_index])
    }

inputs = [
    "I need help generating leads",
    "Looking for an AI chatbot for my website",
    "Want to improve conversion rates",
    "Need help building a sales funnel",
    "Interested in marketing automation tools"
]

query = "I want to get more leads for my business"

result = find_best_match(query, inputs)

print("Query:", result["query"])
print("Best Match:", result["best_match"])
print("Score:", round(result["score"], 3))
```

```python
def get_top_k_matches(query, inputs, k=3):
    input_embeddings = model.encode(inputs)
    query_embedding = model.encode([query])

    scores = cosine_similarity(query_embedding, input_embeddings)[0]

    ranked = sorted(zip(inputs, scores), key=lambda x: x[1], reverse=True)

    return ranked[:k]

top_matches = get_top_k_matches(query, inputs)

for i, (text, score) in enumerate(top_matches, 1):
    print(f"{i}. {text} (Score: {round(score, 3)})")
```

```python
# Optimized with cached embeddings
input_embeddings_cache = model.encode(inputs)

def fast_similarity(query, inputs, cached_embeddings):
    query_embedding = model.encode([query])
    scores = cosine_similarity(query_embedding, cached_embeddings)[0]

    best_index = np.argmax(scores)
    return inputs[best_index], float(scores[best_index])

match, score = fast_similarity(query, inputs, input_embeddings_cache)

print("Fast Match:", match, "| Score:", round(score, 3))
```
